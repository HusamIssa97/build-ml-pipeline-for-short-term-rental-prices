```python
import wandb, pandas as pd

run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
local_path = wandb.use_artifact("sample.csv:latest").file()
df = pd.read_csv(local_path)

df.shape, df.head()
```

```python
df.info()
df.describe(include="all").T
df.isna().sum().sort_values(ascending=False).head(15)
df["price"].describe()
df["last_review"].head(10)
```

```python
# Try ydata-profiling first (modern package name)
try:
    from ydata_profiling import ProfileReport
except ImportError:
    import pandas_profiling as pp
    ProfileReport = pp.ProfileReport

profile = ProfileReport(df, minimal=True, title="NYC Airbnb – EDA")
profile.to_notebook_iframe()
```

```python
df["last_review"] = pd.to_datetime(df["last_review"], errors="coerce")
df["last_review"].dtype, df["last_review"].isna().mean()
```
